```python
import networkx as nx
import matplotlib.pyplot as plt

# Create the graph
G = nx.Graph()

# Adding nodes with names
nodes = ["Andre", "Beverly", "Carol", "Diane", "Ed", "Fernando", "Garth", "Heather", "Ike", "Jane"]
G.add_nodes_from(nodes)

# Adding edges between nodes 
edges = [("Andre", "Beverly"), ("Andre", "Carol"), ("Andre", "Diane"), ("Andre", "Fernando"), ("Andre", "Garth"),
         ("Beverly", "Diane"), ("Beverly", "Ed"), ("Beverly", "Garth"),
         ("Carol", "Diane"), ("Carol", "Fernando"),
         ("Diane", "Ed"), ("Diane", "Fernando"), ("Diane", "Garth"),
         ("Ed", "Garth"),
         ("Fernando", "Garth"), ("Fernando", "Heather"),
         ("Garth", "Heather"),
         ("Heather", "Ike"),
         ("Ike", "Jane")]
G.add_edges_from(edges)

# Draw the graph
pos = nx.spring_layout(G)  # Using spring layout for better visual spacing
nx.draw(G, pos, with_labels=True, node_color='lightblue', edge_color='purple', node_size=500, font_size=10, font_color='black')
plt.title('Krackhardt Kite Graph')
plt.show()

```
